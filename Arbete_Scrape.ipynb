# Looping through 35 pages, get price for audis

```python
bilar_pris = []
for i in range(1, 35):
    
    url = f"https://www.blocket.se/annonser/hela_sverige/fordon/bilar?cb=2&cg=1020&f=p&page={i}"
    html = requests.get(url).content
    key1 = "div.styled__SalesInfo-sc-1kpvi4z-22"
    soup = bs(html)
    
    pris = [i.text for i in soup.select(key1)]
    bilar_pris.append(pris)
    





```

# Bilar namn bli fel

```python
bilar_namn = []
for i in range(1, 35):
    
    url = f"https://www.blocket.se/annonser/hela_sverige/fordon/bilar?cb=2&cg=1020&f=p&page={i}"
    html = requests.get(url).content
    key2 = "span.styled__SubjectContainer-sc-1kpvi4z-11.jzzuDW"
    soup = bs(html)
    
    pris += [i.text for i in soup.select(key2)][0]
    bilar_namn.append(pris)
```
